```python
import csv
import re
import nltk

counter = 1

while (counter <= 426):
    
    def new_text(text):
        # Remove "RT", @ mentions and websites
        text = re.sub(r'RT', '', text)
        text = re.sub(r'@\S+', '', text)
        text = re.sub(r'http\S+', '', text)

        # Remove punctuation and convert to lowercase
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower()

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text)

        return text.strip()

    # Open the CSV file and read the contents
    with open('WEEK {}.csv'.format(counter)) as file:
        reader = csv.reader(file)

        # Skip the first row
        next(reader)

        # Open the new CSV file in write mode
        with open('/WEEK {}.csv'.format(counter), 'w', newline='') as output_file:
            writer = csv.writer(output_file)

            # Write the header row
            writer.writerow(['Cleaned Text'])

            # Iterate over each row in the input CSV file
            for row in reader:
                # The first column in the row is the text
                text = row[0]

                # Clean the text
                clean_text = new_text(text)

                # Write the text and cleaned text to the new CSV file
                writer.writerow([clean_text])
    
    counter += 1
```

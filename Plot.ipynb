```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime
import numpy as np
import pandas as pd

# Open the input CSV file in read mode
with open('output.csv', 'r') as infile:
    reader = csv.reader(infile)

    # Initialize lists to store the dates and values
    dates1 = []
    values1 = []

    # Iterate over the rows in the input CSV file
    for row in reader:
        # The date is in the first column of the row
        date_string = row[0]

        # Convert the date string to a datetime object
        date = datetime.strptime(date_string, '%d/%m/%Y')

        # The value is in the second column of the row
        value_string = row[1]

        # Convert the value string to a float
        value = float(value_string)

        # Add the date and value to the lists
        dates1.append(date)
        values1.append(value)

        # Set the x-axis to the dates and the y-axis to the values
        

# Open the input CSV file in read mode
with open('Sentiment.csv', 'r') as infile:
    reader = csv.reader(infile)

    # Initialize lists to store the dates and values
    dates2 = []
    values2 = []

    # Iterate over the rows in the input CSV file
    for row in reader:
        # The date is in the first column of the row
        date_string = row[0]

        # Convert the date string to a datetime object
        date = datetime.strptime(date_string, '%d/%m/%Y')

        # The value is in the second column of the row
        value_string = row[1]

        # Convert the value string to a float
        value = float(value_string)

        # Add the date and value to the lists
        dates2.append(date)
        values2.append(value)

# Calculate the 10-day moving average for the values from the first CSV file
ma1 = pd.Series(values1).rolling(10).mean()

# Calculate the 30-day moving average for the values from the second CSV file
ma2 = pd.Series(values2).rolling(30).mean()

# Plot the moving averages from the first CSV file
plt.plot(dates1, ma1, color = 'red', label = 'Polling')
plt.plot(dates2, ma2, color = 'black', label = 'Sentiment')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Raw Values')

# Show the plot
plt.show()
```
